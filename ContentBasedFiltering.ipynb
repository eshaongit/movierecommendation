```python
import ast
```

```python
def converter(text):
    List1 = []
    for i in ast.literal_eval(text):
        List1.append(i['name']) 
    return List1 
```

```python
movies_df.dropna(inplace=True)
```

```python
movies_df['genres'] = movies_df['genres'].apply(converter)
movies_df.head()
```

```python
movies_df['keywords'] = movies_df['keywords'].apply(converter)
movies_df.head()
```

```python
movies_df['cast'] = movies_df['cast'].apply(converter)
movies_df.head()
```

```python
movies_df['cast'] = movies_df['cast'].apply(lambda x:x[0:3])
```

```python
def director(text):
    List3 = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            List3.append(i['name'])
    return List3 
```

```python
movies_df['crew'] = movies_df['crew'].apply(director)

movies_df.sample(5)
```

```python
def collapse(List):
    List2 = []
    for i in List:
        List2.append(i.replace(" ",""))
    return List2
```

```python
movies_df['cast'] = movies_df['cast'].apply(collapse)
movies_df['crew'] = movies_df['crew'].apply(collapse)
movies_df['genres'] = movies_df['genres'].apply(collapse)
movies_df['keywords'] = movies_df['keywords'].apply(collapse)
```

```python
movies_df.head(3)
```

```python
movies_df['overview'] = movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df['tags'] = movies_df['overview'] + movies_df['genres'] + movies_df['keywords'] + movies_df['cast'] + movies_df['crew']
```

```python
new = movies_df.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```
